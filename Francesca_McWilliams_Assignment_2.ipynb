```python
def get_businesses(query, api_key, min_rating=0.0, max_results = 30):
    import requests
    import json
    import time
    
    query = query.replace(' ','%20')
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json?input=%s&key=%s" % (query, api_key)
    try:
        response = requests.get(url)
        if not response.status_code == 200: 
            print("HTTP error", response.status_code)
        else:
            try:
                json_dict = response.json()
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")
    
    bus_list = []
    for item in json_dict["results"]:
        if item["rating"] >= min_rating and len(bus_list) < max_results:
            name = item["name"]
            address = item["formatted_address"]
            if "opening_hours" in item: hours = item["opening_hours"]
            else: hours = None
            if "price_level" in item: price = item["price_level"]
            else: price = None
            if "rating" in item: rating = item["rating"]
            else: rating = None
            tup = (name, address, hours, price, rating)
            bus_list.append(tup)
    
    if max_results > 20:
        npt = json_dict["next_page_token"] 
        new_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&key=%s" % (npt, api_key)
        response = requests.get(new_url)
        while response.json().get('status') != 'OK': 
            time.sleep(1)
            response = requests.get(new_url)
            json_dict_2 = response.json()
            
        for item in json_dict_2["results"]:
            if item["rating"] >= min_rating and len(bus_list) < max_results:
                name = item["name"]
                address = item["formatted_address"]
                if "opening_hours" in item: hours = item["opening_hours"]
                else: hours = None
                if "price_level" in item: price = item["price_level"]
                else: price = None
                if "rating" in item: rating = item["rating"]
                else: rating = None
                tup = (name, address, hours, price, rating)
                bus_list.append(tup)
   
    if max_results > 40:
        npt = json_dict_2["next_page_token"]        
        new_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&key=%s" % (npt,api_key)
        response = requests.get(new_url)
        while response.json().get('status') != 'OK': 
            time.sleep(1)
            response = requests.get(new_url)
            json_dict_3 = response.json()
        
        for item in json_dict_3["results"]:
            if item["rating"] >= min_rating and len(bus_list) < max_results:
                name = item["name"]
                address = item["formatted_address"]
                if "opening_hours" in item: hours = item["opening_hours"]
                else: hours = None
                if "price_level" in item: price = item["price_level"]
                else: price = None
                if "rating" in item: rating = item["rating"]
                else: rating = None
                tup = (name, address, hours, price, rating)
                bus_list.append(tup)
    
    return bus_list
```
